```python
df['Total'] = pd.to_numeric(df['Total'],errors='coerce')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scatterplot
sns.scatterplot(data=df,x='Quantity',y='Tax 5%')
```

```python
fig,ax= plt.subplots()
ax.plot(df['Tax 5%'], df['gross income'],ls=':',lw=0.7) # ls is line drawing style , lw is width of line
ax.set_xlabel('tax')
ax.set_ylabel('gross_income')
```

```python
fig,axes= plt.subplots(ncols=2)
df.plot.scatter(x='Tax 5%', y='gross income', ax=axes[0])
df.plot.scatter(x='Quantity', y='cogs', ax=axes[1])
fig.tight_layout
```

```python
# a function to create a histogram
df['Quantity'].plot.hist(bins=30)
```

```python
# a function to create two histogram and concatenate them into one  (multilayer histogram)
fig, ax = plt.subplots()
ax.hist(df['Quantity'], bins=20, histtype='step', label=df['Tax 5%'])
ax.hist(df['Total'], bins=20, histtype='stepfilled', label=df['Payment'], alpha=0.3, edgecolor='w')
ax.legend()
```

```python
#Box plot
df[['gross income', 'Tax 5%']].plot.box();
#Sad :( saari income almost 5% tax dene per laggayi
```

```python
# using this plot with seaborn library
sns.swarmplot(data=df[['Unit price','Quantity']], size=5);
```

```python
plt.pie(df['Quantity'],labels= df['Gender'],autopct = '%.1f%%')
plt.title('Qty by gender')
plt.show()
```
